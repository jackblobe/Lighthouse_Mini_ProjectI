## Task III
#### Locate and list out all bike points within london, including their location, the amount of empty stalls, and the amount of full. 

```python
url = "https://api.tfl.gov.uk/BikePoint"
bikePoint = requests.get(url)
```

```python
bikePoint=bikePoint.json()
print(len(bikePoint))
```

```python
for bikepoint in bikePoint:
    name=bikepoint['commonName']
    docks=bikepoint['additionalProperties'][-1]['value']
    empty=bikepoint['additionalProperties'][-2]['value']
    print(f'Name: {name},Docks:{docks},Empty:{empty}')
```

## Task IV
#### Find the amount of tube and bus lines and print out the tube line names. 

```python
tubeurl='https://api.tfl.gov.uk/line/mode/tube/status' 
tube=requests.get(tubeurl)
tube=tube.json()
len(tube)
print('There are 11 tube lines in London, here are the names:')
for i in range(len(tube)):
    j=(tube[i])
    name=j['name']
    print(f"{i+1}) {name} line")
    
```

```python
busurl='https://api.tfl.gov.uk/line/mode/bus/status'
bus=requests.get(busurl)
bus=bus.json()
print(f"There are {len(bus)} bus lines in London, maybe a few too many to list for you...")
```

## Task VI
#### Plan the journey from Heathrow Airport to Tower Bridge using Public Transport, Taxi or Bike? Which way is the fastest? 

```python
#lat long of heathrow 51.4700,0.4543
#lat long of Tower Bridge 51.055, 0.0754
#both taken from google.com
```

```python
#for next park modes are useful for cycling through journeys
#list good for cycling through loop with given modes
#so update modes from Task II
url = "https://api.tfl.gov.uk/Journey/Meta/Modes"
r = requests.get(url)
json = r.json()
#need to convert modes to 
modes = list([item.get('modeName') for item in json])
print('Planned duration:')
for m in modes:
    if m in ["bus", "tube", "cycle"]:#modes provided in questions
        mode = requests.get(f'https://api.tfl.gov.uk/Journey/JourneyResults/51.4700%2C%20-0.4543/to/51.5055%2C%20-0.0754?mode={m}')
        mode_trip = mode.json()

        if mode.status_code == 200:#put this in after noticing there is an error for cycle

            journey = mode_trip['journeys']
            durations = []
            for j in journey:
                durations.append(j.get('duration'))

            print(f'{m}: {min(durations)} minutes')

        else:
            print(f"{m}: {mode_trip['message']}")#because cycle doesn't show in trip planner so error will occur
```
